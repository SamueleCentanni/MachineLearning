```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

load the data into a dataframe df, show its size and head, eliminate

```python
url = './country_stats.csv'

df = pd.read_csv(url)
df.head()
```

```python
print("The size of the dataframe is {} rows and {} columns".format(df.shape[0], df.shape[1]))
```

```python
null_rows = df.shape[0] - df.dropna().shape[0]
df = df.dropna()

print("The number of rows that have been eliminated is {}".format(null_rows))
print("The new shape is: ", df.shape)

df.head()
```

We can see lot's of outliers from the pairplot. Let's plot the boxplots.

```python
plt.figure(figsize=(15,15))

pos = 1
df0 = df.drop(['Country Name'], axis=1)
for col in df0.columns:
    plt.subplot(4,3,pos)
    sns.boxplot(df0[col])
    pos += 1
    
plt.show()
```

```python
clean_df = df0.drop(columns=['GDP at market prices (current US$)', 'GDP per capita (current US$)'], axis=1)
clean_df.head()
```

Split the reduced data: store the first column in a vector keys and the 
others in a matrix X

```python
X = clean_df
y = df['Country Name']

print("X shape {}, y shape {}".format(X.shape, y.shape))
```

Find the best clustering scheme for the data (possibly reduced after 
step 4) with a method of your choice, plot global silhouette index for 
an appropriate range of hyperparameter(s) and show the chosen 
hyperparameter(s)

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import silhouette_score

params = [{'n_clusters': list(range(2,11))}]
pg = list(ParameterGrid(param_grid=params))

silhouette = []
for i in range(len(pg)):
    model = KMeans(**(pg[i]))
    y_pred = model.fit_predict(X)
    
    silhouette.append(silhouette_score(X, y_pred))
```

```python
plt.plot(list(range(2,11)), silhouette, color='blue')
plt.title("Silhouette score")
plt.xlabel("N clusters")
plt.ylabel("Silhouette score")
#plt.ylim(0,1)
```
